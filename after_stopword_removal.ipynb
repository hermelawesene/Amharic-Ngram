```python
def generate_ngrams(tokens: list, n: int) -> list:
    """
    Generates n-grams from a list of tokens.
    
    Args:
        tokens: A list of words (strings).
        n: The size of the n-gram (e.g., 1 for unigram, 2 for bigram).
        
    Returns:
        A list of n-grams, where each n-gram is a string.
    """
    # Create tuples of length n using a sliding window
    # e.g., for n=3: (t[0], t[1], t[2]), (t[1], t[2], t[3]), ...
    ngrams = zip(*[tokens[i:] for i in range(n)])
    
    # Join the words in each n-gram tuple back into a single string
    return [" ".join(ngram) for ngram in ngrams]
```

```python
def calculate_sentence_probability(sentence_tokens: list, unigram_counter: Counter, bigram_counter: Counter, total_tokens: int):
    """Calculates the probability of a sentence using the Bigram Model (MLE)."""
    
    probability = 1.0
    
    print("\n--- Bigram Calculation Breakdown ---")

    for i in range(len(sentence_tokens)):
        w_i = sentence_tokens[i]
        
        if i == 0:
            # P(w_1) = Count(w_1) / Total Tokens
            w_1_count = unigram_counter.get(w_i, 0)
            
            if w_1_count == 0:
                print(f"P({w_i}) = 0.0 (Word not found in corpus)")
                return 0.0
            
            P_w1 = w_1_count / total_tokens
            probability *= P_w1
            print(f"P({w_i}) = {P_w1:.8f} (Unigram)")
            
        else:
            # P(w_i | w_i-1) = Count(w_i-1 w_i) / Count(w_i-1)
            w_i_minus_1 = sentence_tokens[i-1]
            bigram_key = f"{w_i_minus_1} {w_i}"
            
            bigram_count = bigram_counter.get(bigram_key, 0)
            w_i_minus_1_count = unigram_counter.get(w_i_minus_1, 0)
            
            if bigram_count == 0 or w_i_minus_1_count == 0:
                 # Probability is 0 if the sequence (bigram) or its prefix (unigram) is not found.
                 print(f"P({w_i} | {w_i_minus_1}) = 0.0 (Bigram not found: '{bigram_key}')")
                 return 0.0
            
            P_wi_given_wiminus1 = bigram_count / w_i_minus_1_count
            probability *= P_wi_given_wiminus1
            print(f"P({w_i} | {w_i_minus_1}) = {P_wi_given_wiminus1:.8f} (Bigram)")
            
    return probability
```

```python
import random


def generate_sentence(ngram_counter, n: int, start_word: str, length: int):
    """Generates a random sentence using an N-gram model."""
    
    current_tokens = start_word.split()
    sentence = list(current_tokens)

    if n == 1:
        # Unigram model: choose any word with probability P(w_i)
        words, counts = zip(*ngram_counter.items())
        total_count = sum(counts)
        probabilities = [c / total_count for c in counts]
        
        for _ in range(length - len(current_tokens)):
            next_word = random.choices(words, weights=probabilities, k=1)[0]
            sentence.append(next_word)
        return " ".join(sentence)

    else:
        # Bigram, Trigram, etc. model: choose based on the prefix P(w_i | w_i-n+1...w_i-1)
        
        for _ in range(length - len(current_tokens)):
            # The context is the last (n-1) words
            context = " ".join(sentence[-(n-1):])
            
            # Find all possible next words (suffixes) that follow the current context
            possible_next_words = {}
            total_next_count = 0
            
            # Iterate through all N-grams in the counter
            for ngram, count in ngram_counter.items():
                if ngram.startswith(context + " "):
                    # Extract the word w_i
                    next_word = ngram.split()[-1]
                    possible_next_words[next_word] = count
                    total_next_count += count
            
            if not possible_next_words:
                # If no word follows the context, stop generation
                break

            # Choose the next word based on frequency P(w_i | context)
            words, counts = zip(*possible_next_words.items())
            probabilities = [c / total_next_count for c in counts]
            next_word = random.choices(words, weights=probabilities, k=1)[0]
            
            sentence.append(next_word)
        return " ".join(sentence)
```

```python
import numpy as np
from collections import Counter

# --- Define Test Data (REQUIRED) ---
# NOTE: Use an actual list of tokens from a separate test file.
TEST_TOKENS = test_tokens

# Required variables (assuming they are defined from your corpus):
# unigram_filtered, bigram_filtered, trigram_filtered, total_filtered_tokens

try:
    VOCABULARY_SIZE = len(unigram_train)
except NameError:
    print("ERROR: Please ensure the filtered counters (unigram_filtered, etc.) are defined.")
    exit()

def calculate_perplexity(n: int, test_tokens: list, unigram_c: Counter, bigram_c: Counter, trigram_c: Counter, V: int, total_tokens: int):
    """Calculates Add-1 Smoothed Perplexity for N=1, 2, or 3."""
    
    N_test = len(test_tokens)
    if N_test == 0:
        return float('inf')
        
    P_W_log = 0.0

    for i in range(N_test):
        w_i = test_tokens[i]
        
        if n == 1:
            # P(w_i) = (Count(w_i) + 1) / (Total + V)
            w_i_count = unigram_c.get(w_i, 0)
            P_w_smoothed = (w_i_count + 1) / (total_tokens + V)
            
        elif n == 2:
            # P(w_i | w_i-1) = (Count(w_i-1 w_i) + 1) / (Count(w_i-1) + V)
            if i == 0:
                # Approximate P(w_1) using smoothed unigram probability
                w_i_count = unigram_c.get(w_i, 0)
                P_w_smoothed = (w_i_count + 1) / (total_tokens + V)
            else:
                w_i_minus_1 = test_tokens[i-1]
                bigram_key = f"{w_i_minus_1} {w_i}"
                
                bigram_count = bigram_c.get(bigram_key, 0)
                unigram_count = unigram_c.get(w_i_minus_1, 0)
                
                P_w_smoothed = (bigram_count + 1) / (unigram_count + V)
                
        elif n == 3:
            # P(w_i | w_i-2 w_i-1) = (Count(w_i-2 w_i-1 w_i) + 1) / (Count(w_i-2 w_i-1) + V)
            if i < 2:
                # Use Bigram approximation for the first two words
                if i == 0:
                    w_i_count = unigram_c.get(w_i, 0)
                    P_w_smoothed = (w_i_count + 1) / (total_tokens + V)
                else: # i == 1
                    w_i_minus_1 = test_tokens[i-1]
                    bigram_key = f"{w_i_minus_1} {w_i}"
                    
                    bigram_count = bigram_c.get(bigram_key, 0)
                    unigram_count = unigram_c.get(w_i_minus_1, 0)
                    P_w_smoothed = (bigram_count + 1) / (unigram_count + V)
            else:
                # Use Trigram formula
                w_i_minus_2 = test_tokens[i-2]
                w_i_minus_1 = test_tokens[i-1]
                trigram_key = f"{w_i_minus_2} {w_i_minus_1} {w_i}"
                bigram_prefix_key = f"{w_i_minus_2} {w_i_minus_1}"
                
                trigram_count = trigram_c.get(trigram_key, 0)
                bigram_prefix_count = bigram_c.get(bigram_prefix_key, 0)
                
                P_w_smoothed = (trigram_count + 1) / (bigram_prefix_count + V)
                
        P_W_log += np.log(P_w_smoothed)

    # PPL = (1 / exp(P_W_log)) ^ (1/N)
    #perplex_base = 1 / np.exp(P_W_log)
    #perplexity = np.power(perplex_base, 1/N_test)
    
    # 1. Calculate the average log probability
    average_log_prob = P_W_log / N_test

    # 2. Calculate Perplexity using the stable log-space formula
    # PPL = exp( - (1/N) * SUM(log P) )
    perplexity = np.exp(-average_log_prob)
    return perplexity
```

```python
total_filtered_tokens = total_train_tokens
# --- Execution ---
print("--- Evaluating N-gram Models using Perplexity (Add-1 Smoothing) ---")
print(f"Test Set Size: {len(TEST_TOKENS)} tokens")
print(f"Vocabulary Size (V): {VOCABULARY_SIZE:,} unique words")
print("-" * 65)

results = {}

# N=1 (Unigram)
try:
    ppl_1 = calculate_perplexity(1, TEST_TOKENS, unigram_train, bigram_train, trigram_train, VOCABULARY_SIZE, total_filtered_tokens)
    results[1] = ppl_1
    print(f"Perplexity (N=1 Unigram): {ppl_1:.2f}")
except Exception as e:
    print(f"N=1 Calculation Error: {e}")

# N=2 (Bigram)
try:
    ppl_2 = calculate_perplexity(2, TEST_TOKENS, unigram_train, bigram_train, trigram_train, VOCABULARY_SIZE, total_filtered_tokens)
    results[2] = ppl_2
    print(f"Perplexity (N=2 Bigram):  {ppl_2:.2f}")
except Exception as e:
    print(f"N=2 Calculation Error: {e}")

# N=3 (Trigram)
try:
    ppl_3 = calculate_perplexity(3, TEST_TOKENS, unigram_train, bigram_train, trigram_train, VOCABULARY_SIZE, total_filtered_tokens)
    results[3] = ppl_3
    print(f"Perplexity (N=3 Trigram): {ppl_3:.2f}")
except Exception as e:
    print(f"N=3 Calculation Error: {e}")

print("-" * 65)

# --- Final Conclusion ---
if results:
    best_n = min(results, key=results.get)
    print(f"The best model is the N={best_n} model with a Perplexity of {results[best_n]:.2f}.")
```

```python
import numpy as np

# --- 1. Define Candidate Words and Test Cases ---
# Candidate words for the missing slot (W_candidate)
CANDIDATE_WORDS = ["ኢትዮጵያ", "አፍሪካ", "ወደ", "ሀገር", "መኖር", "ነው", "ግን", "በ", "አንድ"] 

# Test cases: (Context_Prefix, Missing_Word_Index, Correct_Prediction)
# The index is the position of the missing word slot (0-indexed)
# The correct prediction must be a word that exists in CANDIDATE_WORDS
TEST_SENTENCES = [
    (["በደቡብ", "___", "ተካሂዷል"], 1, "አፍሪካ"),  # "በደቡብ አፍሪካ ተካሂዷል" (Trigram test)
    (["ልጆች", "___", "ይሄዳሉ"], 1, "ወደ"),      # "ልጆች ወደ ይሄዳሉ" (Bigram test, requires context 'ልጆች')
    (["አንዲት", "ቆንጆ", "___"], 2, "ሀገር"),       # "አንዲት ቆንጆ ሀገር" (Trigram test)
    (["ብራዘርስ", "___"], 1, "አፍሪካ"),            # "ብራዘርስ አፍሪካ" (Bigram test)
]

def predict_next_word(context_prefix: list, candidates: list, n: int, model_c: Counter, prefix_c: Counter, V: int) -> str:
    """Predicts the best candidate word to complete the context using Add-1 Smoothing."""
    
    prefix = " ".join(context_prefix)
    best_candidate = ""
    max_log_prob = -np.inf

    for w_i in candidates:
        if n == 2:
            # Context is w_i-1 (the last word of the prefix)
            w_i_minus_1 = context_prefix[-1]
            bigram_key = f"{w_i_minus_1} {w_i}"
            
            # P(w_i | w_i-1) = (Count(w_i-1 w_i) + 1) / (Count(w_i-1) + V)
            bigram_count = model_c.get(bigram_key, 0)
            unigram_count = prefix_c.get(w_i_minus_1, 0)
            
            # The prefix_c here is unigram_train
            P_w_smoothed = (bigram_count + 1) / (unigram_count + V)
            
        elif n == 3:
            # Context is w_i-2 w_i-1 (the last two words of the prefix)
            w_i_minus_2 = context_prefix[-2]
            w_i_minus_1 = context_prefix[-1]
            trigram_key = f"{w_i_minus_2} {w_i_minus_1} {w_i}"
            bigram_prefix_key = f"{w_i_minus_2} {w_i_minus_1}"
            
            # P(w_i | w_i-2 w_i-1) = (Count(w_i-2 w_i-1 w_i) + 1) / (Count(w_i-2 w_i-1) + V)
            trigram_count = model_c.get(trigram_key, 0)
            # The prefix_c here is bigram_train
            bigram_prefix_count = prefix_c.get(bigram_prefix_key, 0)
            
            P_w_smoothed = (trigram_count + 1) / (bigram_prefix_count + V)
        else:
            return "N must be 2 or 3"
        
        # We maximize the log probability
        log_prob = np.log(P_w_smoothed)
        
        if log_prob > max_log_prob:
            max_log_prob = log_prob
            best_candidate = w_i
            
    return best_candidate

# --- 2. Evaluate Models ---
def evaluate_extrinsic(model_n: int, model_c: Counter, prefix_c: Counter, V: int, title: str):
    """Runs the extrinsic evaluation and prints accuracy."""
    
    correct_predictions = 0
    total_tests = len(TEST_SENTENCES)
    
    print(f"\n--- {title} Extrinsic Evaluation (N={model_n}) ---")
    
    for prefix_list, missing_idx, correct_word in TEST_SENTENCES:
        # Create the context for prediction: exclude the placeholder
        context = [w for i, w in enumerate(prefix_list) if w != '___']
        
        # Handle cases where the required prefix length is not met
        if (model_n == 3 and len(context) < 2) or (model_n == 2 and len(context) < 1):
            # Skip tests that don't fit the N-gram size
            continue
        
        predicted_word = predict_next_word(context, CANDIDATE_WORDS, model_n, model_c, prefix_c, V)
        
        if predicted_word == correct_word:
            correct_predictions += 1
            result = "CORRECT"
        else:
            result = f"WRONG (Predicted: {predicted_word})"

        full_phrase = prefix_list[:missing_idx] + [predicted_word] + prefix_list[missing_idx+1:]
        print(f"Test: {' '.join(prefix_list)} -> Prediction: {' '.join(full_phrase)} | {result}")
        
    accuracy = (correct_predictions / total_tests) * 100
    print("-" * 60)
    print(f"Final Accuracy ({title}): {accuracy:.2f}%")
    return accuracy

# --- 3. Execute Evaluation ---


try:
    # V is vocabulary_size, total_train_tokens is the total count of training tokens.
    # The prefix_c for Bigram is the unigram_train counter.
    # The prefix_c for Trigram is the bigram_train counter.
    
    # Evaluate Bigram (N=2)
    bigram_accuracy = evaluate_extrinsic(
        model_n=2, 
        model_c=bigram_train, 
        prefix_c=unigram_train, # P(w_i-1) is a unigram count
        V=vocabulary_size, 
        title="Bigram Model"
    )

    # Evaluate Trigram (N=3)
    trigram_accuracy = evaluate_extrinsic(
        model_n=3, 
        model_c=trigram_train, 
        prefix_c=bigram_train, # P(w_i-2 w_i-1) is a bigram count
        V=vocabulary_size, 
        title="Trigram Model"
    )

except NameError as e:
    print(f"\nEvaluation Error: {e}. Please ensure bigram_train, trigram_train, unigram_train, and vocabulary_size are correctly defined from your training set.")
```
