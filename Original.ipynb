```python
from collections import Counter

# ---- Function to build n-gram counts ----
def get_ngram_counts(tokens, n):
    ngrams = zip(*[tokens[i:] for i in range(n)])
    return Counter(ngrams)

# ---- Function to compute probabilities ----
def ngram_probabilities(counts):
    total = sum(counts.values())
    return {ng: count / total for ng, count in counts.items()}
```

```python
def generate_ngrams(tokens: list, n: int) -> list:
    """
    Generates n-grams from a list of tokens.
    
    Args:
        tokens: A list of words (strings).
        n: The size of the n-gram (e.g., 1 for unigram, 2 for bigram).
        
    Returns:
        A list of n-grams, where each n-gram is a string.
    """
    # Create tuples of length n using a sliding window
    # e.g., for n=3: (t[0], t[1], t[2]), (t[1], t[2], t[3]), ...
    ngrams = zip(*[tokens[i:] for i in range(n)])
    
    # Join the words in each n-gram tuple back into a single string
    return [" ".join(ngram) for ngram in ngrams]
```

```python
def ngram_probabilities(counter):
    total = sum(counter.values())
    return {ng: count / total for ng, count in counter.items()}
```

```python
def show_top(counter, top=20):
    probs = ngram_probabilities(counter)
    
    # sort by count (descending)
    top_items = counter.most_common(top)

    for ng, count in top_items:
        print(f"{' '.join(ng)}  ->  count={count:,}   prob={probs[ng]:.6f}")
```

```python
def bigram_conditional_probability(bigram_counts, unigram_counts, w1, w2):
    """Compute P(w2 | w1)."""
    return bigram_counts.get((w1, w2), 0) / unigram_counts.get((w1,), 1)
```

```python
def calculate_bigram_conditional_probability(bigram_counter, unigram_counter, top_k=10):
    """
    Calculates P(w_i | w_i-1) for the top bigrams.
    
    Args:
        bigram_counter (Counter): Counter of 'w_i-1 w_i' strings.
        unigram_counter (Counter): Counter of 'w_i-1' strings (the prefix).
        top_k (int): Number of top bigrams to display.
    """
    print("\n" + "="*70)
    print(f"## Conditional Probability P(w_i | w_i-1) for Top {top_k} Bigrams")
    print("="*70)
    
    # Get the top K bigrams sorted by their count
    top_bigrams = bigram_counter.most_common(top_k)

    # Print header
    print(f"{'Bigram (w_i-1 w_i)':<30} | {'Count(w_i-1, w_i)':<18} | {'Count(w_i-1)':<12} | {'P(w_i | w_i-1)':<15}")
    print("-" * 70)

    for bigram_str, bigram_count in top_bigrams:
        # Split the bigram string into prefix (w_i-1) and current word (w_i)
        w_i_minus_1 = bigram_str.split()[0]
        
        # Get the count of the prefix word (the denominator)
        w_i_minus_1_count = unigram_counter.get(w_i_minus_1, 0)
        
        # Calculate the conditional probability
        if w_i_minus_1_count > 0:
            conditional_prob = bigram_count / w_i_minus_1_count
        else:
            conditional_prob = 0.0 # Should not happen if data is consistent

        prob_formatted = f"{conditional_prob:.6f}"
        
        # Display the result
        print(f"{bigram_str:<30} | {bigram_count:<18} | {w_i_minus_1_count:<12} | {prob_formatted:<15}")

# Execute the calculation and display
calculate_bigram_conditional_probability(bigram, unigram, top_k=10)
```

```python
# --- 3. N-Gram Calculation and Display ---

for n in N_RANGE:
    # 1. Generate N-Grams using the filtered tokens
    ngrams = (filtered_tokens, n=n)
    
    # 2. Count frequencies
    counts = Counter(ngrams)
    total_ngrams = len(ngrams)
    
    # 3. Get top N-grams
    sorted_counts = counts.most_common(TOP_K)
    
    print(f"\n## üìà Top {TOP_K} {n}-grams (AFTER STOPWORD REMOVAL)")
    
    # Print header
    print(f"{'N-gram':<35} | {'Count':<6} | {'Probability':<12}")
    print("-" * 70)
    
    # Iterate and display results
    for ngram, count in sorted_counts:
        probability = count / total_ngrams
        prob_formatted = f"{probability:.5f}"
        
        print(f"{ngram:<35} | {count:<6} | {prob_formatted:<12}")

print("\n" + "="*70)
print("Stopword removal and n-gram analysis complete.")
```
